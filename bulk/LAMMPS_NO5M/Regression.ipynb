```python
# 15 m/s X

import re
import matplotlib.pyplot as plt

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(6)

loc = './15/X/'
file0 = 'NO5M'
file = file0 + '_deform_15' + '.dat'
f = open(loc+file, 'r')
stress = [0.0]
strain = [0.0]
per20 = 0
curverange = [0.0, 3.5]
if curverange[0] == 0.0:
    strain_per20 = [0.0]
    per20_box = [0.0]
else:
    strain_per20 = []
    per20_box = []
for i,line in enumerate(f):
    if i >= 2:
        a = re.findall(r'\-?[0-9]+\.?[0-9]+', line)
        if eval(a[2]) / 43.4470 - 1 >= curverange[0] and eval(a[2]) / 43.4470 - 1 <= curverange[1]:
            strain.append((eval(a[2]) / 43.4470 - 1)*100)
            stress.append(-eval(a[-1])/10.0)
            #print(-eval(a[-1])/10.0)
            per20 = per20 - eval(a[-1])/10.0
            if i % 20 == 1 and i != 1:
                per20 = per20 / 20
                per20_box.append(per20)
                strain_per20.append((eval(a[2]) / 43.4470 - 1)*100)
                #print('Ave:', per20)
                per20 = 0

f.close()
#plt.plot(strain,stress,'.')
plt.plot(strain_per20, per20_box,'-')
plt.xlabel('Strain(%)')
plt.ylabel('Stress(MPa)')
plt.title('{0} Stress-Strain Curve'.format(file0))
#plt.title('Stress-Strain Curve')
#plt.savefig(filename+'_x.svg', format='svg', bbox_inches='tight', dpi=350)
plt.savefig('1_f.svg', format='svg', bbox_inches='tight', dpi=350)
plt.show()
```

```python
import re
import matplotlib.pyplot as plt
import numpy as np

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(6)

locs = ['./15/X/', './15/Y/', './15/Z/']

lx = 43.4470
ly = 37.1978
lz = 49.5970
ls = [lx, ly, lz]

filename = 'NO5M'
file0 = filename + '_deform_15' + '.dat'

for j in range(len(locs)):
    file = locs[j] + file0
    f = open(file, 'r')
    stress = []
    strain = []
    strain_per10 = [0]
    per10 = 0
    per10_box = [0]
    for i,line in enumerate(f):
        if i >= 2:
            a = re.findall(r'\-?[0-9]+\.?[0-9]+', line)
            if eval(a[j+2]) / ls[j] - 1 >= curverange[0] and eval(a[j+2]) / ls[j] - 1 <= curverange[1]:
                strain.append((eval(a[j+2]) / ls[j] - 1)*100)
                stress.append(-eval(a[-1])/10.0)
                per10 = per10 - eval(a[-1])/10.0

                if i % 10 == 1 and i != 1:
                    per10 = per10 / 10
                    per10_box.append(per10)
                    strain_per10.append((eval(a[j+2]) / ls[j] - 1)*100)
                    per10 = 0
    # Calculate equation for trendline
    z = np.polyfit(strain_per10, per10_box, 7)
    p = np.poly1d(z)
    plt.plot(strain_per10, p(strain_per10))#(*, coleur[j]+'-')
    f.close()

for j in range(len(locs)):
    file = locs[j] + file0
    f = open(file, 'r')
    stress = []
    strain = []
    strain_per5 = [0]
    strain_per50 = [0]
    per5 = 0
    per5_box = [0]
    per50_box = [0]
    per50_inter = [0]
    per50_box_min = [0]
    per50_box_max = [0]
    for i,line in enumerate(f):
        if i >= 2:
            a = re.findall(r'\-?[0-9]+\.?[0-9]+', line)
            if eval(a[j+2]) / ls[j] - 1 >= curverange[0] and eval(a[j+2]) / ls[j] - 1 <= curverange[1]:
                strain.append((eval(a[j+2]) / ls[j] - 1)*100)
                stress.append(-eval(a[-1])/10.0)
                per5 = per5 - eval(a[-1])/10.0
                if i % 5 == 1 and i != 1:
                    per5 = per5 / 5
                    per5_box.append(per5)
                    strain_per5.append((eval(a[j+2]) / ls[j] - 1)*100)
                    per50_inter.append(per5)
                    per5 = 0
                if i % 50 == 1 and i != 1:
                    strain_per50.append((eval(a[j+2]) / ls[j] - 1)*100)
                    per50_box_min.append(min(per50_inter))
                    per50_box_max.append(max(per50_inter))
                    per50_inter = []
    z_min = np.polyfit(strain_per50, per50_box_min, 7)
    p_min = np.poly1d(z_min)
    z_max = np.polyfit(strain_per50, per50_box_max, 7)
    p_max = np.poly1d(z_max)
    plt.fill_between(strain_per50, p_min(strain_per50), p_max(strain_per50), alpha=0.2)
    f.close()

plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'font.family': 'Arial'})
plt.xlabel('Strain(%)')
plt.ylabel('Stress(MPa)')
plt.xlim(-0.01, 310)
plt.ylim(0, 4000)
plt.title('NO5M Stress-Strain Curve'.format(file0))
plt.legend(['X','Y','Z'])
plt.savefig(filename+'_4.svg', format='svg', bbox_inches='tight', dpi=350)
plt.show()
```
