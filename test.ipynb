```python
def size(pos,xyz,elements=None,omit=False):
    min=50
    max=-50
    for i in range(len(pos)):
        if elements!=None:
            if elements[i]=='H':
                continue
        if pos[i][xyz]>max:
            max=pos[i][xyz]
        elif pos[i][xyz]<min:
            min=pos[i][xyz]
    if omit==True:
        pass
    else:
        if xyz==0:
            max=max+2
        else:
            min=min-1
            max=max+1
    return [min,max]

def semi_BADI(atoms,readfile,writefile,not_write=0,infinite=0):
    atoms_number=len(atoms)
    pos=atoms.get_positions()
    i=0
    j=0
    al=[]
    masses=[]
    masstype=[]
    pairs=[]
    pairtype=[]
    ma_irs=[]
    ma_irtype=[]
    bonds=[]
    bondtype=[]
    angles=[]
    angletype=[]
    dihedrals=[]
    dihedraltype=[]
    impropers=[]
    impropertype=[]
    error=0
    cut=0
    for line in readfile:
        #Mass
        a=re.search('Masses',line)
        if a:
            i=1
        if i==1:
            #print(line)
            mass=re.findall(r'[0-9]+\.[0-9]+',line)
            try:
                masses.append(mass[-1])
            except:
                error=error+1
                if error==3:
                    i=0
                    error=0
                continue
            if mass[-1] not in masstype:
                masstype.append(mass[-1])
            #Normal
            if len(masses)==atoms_number:
                i=0
                error=0
        
        #Pairs
        b=re.search('Pair Coeffs',line)
        if b:
            i=2
        if i==2:
            pair=re.findall(r'[0-9]+\.[0-9]+',line)
            try:
                pairs.append([pair[-2],pair[-1]])
            except:
                error=error+1
                if error==3:
                    i=0
                    error=0
                continue
            if [pair[-2],pair[-1]] not in pairtype:
                pairtype.append([pair[-2],pair[-1]])
            if len(masses)==len(pairs):
                for i in range(len(masses)):
                    ma_irs.append([masses[i],pairs[i][-2],pairs[i][-1]])
                    if [masses[i],pairs[i][-2],pairs[i][-1]] not in ma_irtype:
                        ma_irtype.append([masses[i],pairs[i][-2],pairs[i][-1]])
                if not_write==0:
                    writefile.write('Masses\n\n')
                    for k,mp in enumerate(ma_irtype):
                        writefile.write('{0} {1}\n'.format(k+1,mp[0]))
                    writefile.write('\n')
                    writefile.write('Pair Coeffs\n\n')
                    for k,mp in enumerate(ma_irtype):
                        writefile.write('{0} {1:>8} {2:>8}\n'.format(k+1,mp[1],mp[2]))
                    writefile.write('\n')
                i=0
                error=0
                continue

        #Bonds
        c=re.search('Bond Coeffs',line)
        if c:
            i=3
        if i==3:
            bond=re.findall(r'[0-9]+\.[0-9]+',line)
            try:
                bonds.append([bond[-2],bond[-1]])
            except:
                error=error+1
                if error==3:
                    if not_write==0:
                        writefile.write('Bond Coeffs\n\n')
                        for k,bp in enumerate(bondtype):
                            writefile.write('{0} {1:>8} {2:>8}\n'.format(k+1,bp[0],bp[1]))
                        writefile.write('\n')
                    i=0
                    error=0
                continue
            if [bond[-2],bond[-1]] not in bondtype:
                bondtype.append([bond[-2],bond[-1]])

        #Angles
        d=re.search('Angle Coeffs',line)
        if d:
            i=4
        if i==4:
            angle=re.findall(r'[0-9]+\.[0-9]+',line)
            try:
                angles.append([angle[-2],angle[-1]])
            except:
                error=error+1
                if error==3:
                    if not_write==0:
                        writefile.write('Angle Coeffs\n\n')
                        for k,ap in enumerate(angletype):
                            writefile.write('{0} {1:>8} {2:>8}\n'.format(k+1,ap[0],ap[1]))
                        writefile.write('\n')
                    i=0
                    error=0
                continue
            if [angle[-2],angle[-1]] not in angletype:
                angletype.append([angle[-2],angle[-1]])

        #Dihedrals
        e=re.search('Dihedral Coeffs',line)
        if e:
            i=5
        if i==5:
            dihedral=re.findall(r'-?[0-9]+\.[0-9]+',line)
            try:
                dihedrals.append([dihedral[-4],dihedral[-3],dihedral[-2],dihedral[-1]])
            except:
                error=error+1
                if error==3:
                    if not_write==0:
                        writefile.write('Dihedral Coeffs\n\n')
                        for k,dp in enumerate(dihedraltype):
                            writefile.write('{0} {1:>8} {2:>8} {3:>8} {4:>8}\n'.format(k+1,dp[0],dp[1],dp[2],dp[3]))
                        writefile.write('\n')
                    i=0
                    error=0
                continue
            if [dihedral[-4],dihedral[-3],dihedral[-2],dihedral[-1]] not in dihedraltype:
                dihedraltype.append([dihedral[-4],dihedral[-3],dihedral[-2],dihedral[-1]])

        #Impropers
        e=re.search('Improper Coeffs',line)
        if e:
            i=11
        if i==11:
            improper=re.findall(r'-?[0-9]+\.?[0-9]*',line)
            try:
                impropers.append([improper[-3],improper[-2],improper[-1]])
            except:
                error=error+1
                if error==3:
                    if not_write==0:
                        writefile.write('Improper Coeffs\n\n')
                        for k,ip in enumerate(impropertype):
                            writefile.write('{0} {1:>8} {2:>3} {3:>3}\n'.format(k+1,ip[0],ip[1],ip[2]))
                        writefile.write('\n')
                    i=0
                    error=0
                continue
            if [improper[-3],improper[-2],improper[-1]] not in impropertype:
                impropertype.append([improper[-3],improper[-2],improper[-1]])

        #Atoms
        g=re.search('Atoms',line)
        if g:
            i=6
            if not_write==0:
                writefile.write('Atoms\n\n')
        if i==6:
            #Specialized for infinite straight chain
            if infinite==1:
                x1=re.search(' {0} '.format(str(atoms_number-1)),line)
                if x1:
                    continue
                x2=re.search(' {0} '.format(str(atoms_number)),line)
                if x2:
                    continue
            al=re.findall(r'-?[0-9]+\.?[0-9]*',line)
            if al==[]:
                error=error+1
                if error==3:
                    i=0
                    if not_write==0:
                        writefile.write('\n')
                    error=0
                    j=0
                continue
            for k in range(len(ma_irtype)):
                if ma_irs[j]==ma_irtype[k]:
                    if not_write==0:
                        writefile.write('{0} {1} {2} {6:>7.4f} {3:>14.7f} {4:>14.7f} {5:>14.7f}\n'.format(j+1,al[1],k+1,eval(al[-3]),eval(al[-2]),eval(al[-1]),eval(al[-4])))
            j=j+1

        #Bonds
        h=re.search('Bonds',line)
        if h:
            i=7
            if not_write==0:
                writefile.write('Bonds\n\n')
        if i==7:
            #Specialized for infinite straight chain
            if infinite==1:
                x1=re.search(' {0}'.format(str(atoms_number+1)),line)
                if x1:
                    y1=re.findall(r'[0-9]+',line)
                    if eval(y1[0])==atoms_number+1 and eval(y1[-2])!=atoms_number+1 and eval(y1[-1])!=atoms_number+1:
                        pass
                    else:
                        del bonds[eval(y1[0])-1]
                        continue
                x2=re.search(' {0}'.format(str(atoms_number+2)),line)
                if x2:
                    y2=re.findall(r'[0-9]+',line)
                    if eval(y2[0])==atoms_number+2 and eval(y2[-2])!=atoms_number+2 and eval(y2[-1])!=atoms_number+2:
                        pass
                    else:
                        del bonds[eval(y2[0])-1]
                        continue
            bondxyz=re.findall(r'[0-9]+',line)
            if bondxyz==[]:
                error=error+1
                if error==3:
                    i=0
                    if infinite==1:
                        if not_write==0:
                            for k,ty in enumerate(bondtype):
                                print('{0}: {1}'.format(k+1,ty))
                            extra1=input("Input the style of the bond:")
                            extra2=input("Input the first atom numbers(No 0!):")
                            extra3=input("Input the second atom numbers:")
                            writefile.write('{0} {1} {2} {3}\n'.format(j+1,extra1,extra2,extra3))
                            bonds.append([extra2,extra3])
                    if not_write==0:
                        writefile.write('\n')
                    error=0
                    j=0
                continue
            for k in range(len(bondtype)):
                if bonds[j]==bondtype[k]:
                    if not_write==0:
                        writefile.write('{0} {1} {2} {3}\n'.format(j+1,k+1,bondxyz[-2],bondxyz[-1]))
            j=j+1

        #Angles
        m=re.search('Angles',line)
        if m:
            i=8
            if not_write==0:
                writefile.write('Angles\n\n')
        if i==8:
            #Specialized for infinite straight chain
            if infinite==1:
                x1=re.search(' {0}'.format(str(atoms_number+1)),line)
                if x1:
                    y1=re.findall(r'[0-9]+',line)
                    if eval(y1[0])==atoms_number+1 and eval(y1[-3])!=atoms_number+1 and eval(y1[-2])!=atoms_number+1 and eval(y1[-1])!=atoms_number+1:
                        pass
                    else:
                        del angles[eval(y1[0])-1]
                        continue
                x2=re.search(' {0}'.format(str(atoms_number+2)),line)
                if x2:
                    y2=re.findall(r'[0-9]+',line)
                    if eval(y2[0])==atoms_number+2 and eval(y2[-3])!=atoms_number+2 and eval(y2[-2])!=atoms_number+2 and eval(y2[-1])!=atoms_number+2:
                        pass
                    else:
                        del angles[eval(y2[0])-1]
                        continue
            anglexyz=re.findall(r'[0-9]+',line)
            if anglexyz==[]:
                error=error+1
                if error==3:
                    i=0
                    if not_write==0:
                        writefile.write('\n')
                    error=0
                    j=0
                continue
            for k in range(len(angletype)):
                if angles[j]==angletype[k]:
                    if not_write==0:
                        writefile.write('{0} {1} {2} {3} {4}\n'.format(j+1,k+1,anglexyz[-3],anglexyz[-2],anglexyz[-1]))
            j=j+1


        #Dihedrals
        n=re.search('Dihedrals',line)
        if n:
            i=9
            if not_write==0:
                writefile.write('Dihedrals\n\n')
        if i==9:
            #Specialized for infinite straight chain
            if infinite==1:
                x1=re.search(' {0}'.format(str(atoms_number+1)),line)
                if x1:
                    y1=re.findall(r'[0-9]+',line)
                    if eval(y1[0])==atoms_number+1 and eval(y1[-4])!=atoms_number+1 and eval(y1[-3])!=atoms_number+1 and eval(y1[-2])!=atoms_number+1 and eval(y1[-1])!=atoms_number+1:
                        pass
                    else:
                        del dihedrals[eval(y1[0])-1]
                        continue
                x2=re.search(' {0}'.format(str(atoms_number+2)),line)
                if x2:
                    y2=re.findall(r'[0-9]+',line)
                    if eval(y2[0])==atoms_number+2 and eval(y2[-4])!=atoms_number+2 and eval(y2[-3])!=atoms_number+2 and eval(y2[-2])!=atoms_number+2 and eval(y2[-1])!=atoms_number+2:
                        pass
                    else:
                        del dihedrals[eval(y2[0])-1]
                        continue
            dihedralxyz=re.findall(r'[0-9]+',line)
            if dihedralxyz==[]:
                error=error+1
                if error==3:
                    i=0
                    if not_write==0:
                        writefile.write('\n')
                    error=0
                    j=0
                continue
            for k in range(len(dihedraltype)):
                if dihedrals[j]==dihedraltype[k]:
                    if not_write==0:
                        writefile.write('{0} {1} {2} {3} {4} {5}\n'.format(j+1,k+1,dihedralxyz[-4],dihedralxyz[-3],dihedralxyz[-2],dihedralxyz[-1]))
            j=j+1

        #Impropers
        o=re.search('Impropers',line)
        if o:
            i=10
            if not_write==0:
                writefile.write('Impropers\n\n')
        if i==10:
            #Specialized for infinite straight chain
            if infinite==1:
                x1=re.search(' {0}'.format(str(atoms_number+1)),line)
                if x1:
                    y1=re.findall(r'[0-9]+',line)
                    if eval(y1[0])==atoms_number+1 and eval(y1[-4])!=atoms_number+1 and eval(y1[-3])!=atoms_number+1 and eval(y1[-2])!=atoms_number+1 and eval(y1[-1])!=atoms_number+1:
                        pass
                    else:
                        del impropers[eval(y1[0])-1]
                        continue
                x2=re.search(' {0}'.format(str(atoms_number+2)),line)
                if x2:
                    y2=re.findall(r'[0-9]+',line)
                    if eval(y2[0])==atoms_number+2 and eval(y2[-4])!=atoms_number+2 and eval(y2[-3])!=atoms_number+2 and eval(y2[-2])!=atoms_number+2 and eval(y2[-1])!=atoms_number+2:
                        pass
                    else:
                        del impropers[eval(y2[0])-1]
                        continue
            improperxyz=re.findall(r'[0-9]+',line)
            if improperxyz==[]:
                error=error+1
                if error==3:
                    i=0
                    if not_write==0:
                        writefile.write('\n')
                    error=0
                    j=0
                continue
            #print(line)
            for k in range(len(impropertype)):
                if impropers[j]==impropertype[k]:
                    if not_write==0:
                        writefile.write('{0} {1} {2} {3} {4} {5}\n'.format(j+1,k+1,improperxyz[-4],improperxyz[-3],improperxyz[-2],improperxyz[-1]))
            j=j+1

    return bonds,angles,dihedrals,impropers,ma_irtype,bondtype,angletype,dihedraltype,impropertype

#To get Files

# LAMMPS.data
import math as m
import numpy as np
from ase.io import read, write
from ase import Atoms
from ase.visualize import view
import re

filename=input('File:')
atoms=read('{0}.xyz'.format(filename))
atoms_number=len(atoms)
readfile = open('{0}.py'.format(filename),'r')
try:
    f = open('{0}.data'.format(filename),'x')
except:
    f = open('{0}.data'.format(filename),'w')

pos=atoms.get_positions()
infinite=0

#Turn the chain
pos=atoms.get_positions()
Apos=pos[0]
Bpos=pos[atoms_number-5]
atoms.rotate(m.atan((Bpos[-1]-Apos[-1])/(Bpos[1]-Apos[1]))/m.pi*180,'x',Apos)
pos=atoms.get_positions()
Apos=pos[0]
Bpos=pos[atoms_number-5]
atoms.rotate(m.atan((Bpos[-1]-Apos[-1])/(Bpos[0]-Apos[0]))/m.pi*180,'y',Apos)
pos=atoms.get_positions()
Apos=pos[0]
Bpos=pos[atoms_number-5]
atoms.rotate(m.atan(-(Bpos[1]-Apos[1])/(Bpos[0]-Apos[0]))/m.pi*180+180,'z',Apos)
pos=atoms.get_positions()
write('{0}.xyz'.format(filename), atoms)
atoms_number=len(atoms)

elements=atoms.get_chemical_symbols()
elementsIn=['H']
for element in elements:
    if element not in elementsIn:
        elementsIn.append(element)

bonds,angles,dihedrals,impropers,pairtype,bondtype,angletype,dihedraltype,impropertype=semi_BADI(atoms,readfile,f,1,infinite=infinite)
readfile.close()
# Size of the system cell in Å
sizex=size(pos,0)
sizey=size(pos,1)
sizez=size(pos,2)

# First line: Comment Line
f.write('#Polymer: {0}\n\n'.format(filename))

# Number of atoms
f.write('{0} atoms\n'.format(atoms_number))
if infinite==1:
    f.write('{0} bonds\n'.format(len(bonds)+1))
else:
    f.write('{0} bonds\n'.format(len(bonds)))
f.write('{0} angles\n'.format(len(angles)))
f.write('{0} dihedrals\n'.format(len(dihedrals)))
f.write('{0} impropers\n\n'.format(len(impropers)))

f.write('{0} atom types\n'.format(len(pairtype)))
f.write('{0} bond types\n'.format(len(bondtype)))
f.write('{0} angle types\n'.format(len(angletype)))
f.write('{0} dihedral types\n'.format(len(dihedraltype)))
f.write('{0} improper types\n\n'.format(len(impropertype)))

# Box of dimensions
f.write('{0:>10.5f} {1:>10.5f} xlo xhi\n'.format(*sizex))# lo:low, hi:high
f.write('{0:>10.5f} {1:>10.5f} ylo yhi\n'.format(*sizey))
f.write('{0:>10.5f} {1:>10.5f} zlo zhi\n'.format(*sizez))
# 3 dimensions
f.write('\n')

if infinite==0:
    readfile = open('{0}.py'.format(filename),'r')
elif infinite==1:
    readfile = open('{0}.py'.format(filename[:-6]),'r')
x=semi_BADI(atoms,readfile,f,infinite=infinite)

f.close()
```

```python
import py3Dmol
from rdkit import Chem
from rdkit.Chem import AllChem
from ipywidgets import interact

def smi2conf(smiles):
    '''Convert SMILES to rdkit.Mol with 3D coordinates'''
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        mol = Chem.AddHs(mol)
        AllChem.EmbedMolecule(mol)
        return mol
    else:
        return None

def MolTo3DView(mol,size=(640,480),surface=False,opacity=0.5):
    """Draw molecule in 3D (Not Origin)
    
    Args:
    ----
        mol: rdMol, molecule to show
        size: tuple(int, int), canvas size
        style: str, type of drawing molecule
               style can be 'line', 'stick', 'sphere', 'cartoon'
        surface, bool, display SAS
        opacity, float, opacity of surface, range 0.0-1.0
    Return:
    ----
        viewer: py3Dmol.view, a class for constructing embedded 3Dmol.js views in ipython notebooks.
    """
    mblock = Chem.MolToMolBlock(mol)
    viewer = py3Dmol.view(width=size[0],height=size[1])
    viewer.addModel(mblock,'mol')
    viewer.setStyle({'sphere':{'radius':0.4},'stick':{'radius':0.1}})
    if surface:
        viewer.addSurface(py3Dmol.SAS, {'opacity': opacity})
    viewer.zoomTo()
    Chem.MolToXYZFile(mol,'Mole_Untitled.xyz')
    return viewer

@interact
def smi2viewer(smi='CC'):
    try:
        conf = smi2conf(smi)
        
        return MolTo3DView(conf).show()
    except:
        return None
```

```python
from ase.io import read
from ase import Atoms
import numpy as np

def HBondNum(file, HBond=2.0):
    atoms = read(file+'.xyz')
    atoms_H_bond = []
    atoms_H = []
    angle_group = []
    
    for i, atom in enumerate(atoms):
        if atom.symbol == 'O':
            atoms_H_bond.append(i)
        elif atom.symbol == 'F':
            atoms_H_bond.append(i)
        elif atom.symbol == 'N':
            atoms_H_bond.append(i)
        if atom.symbol == 'H':
            atoms_H.append(i)

    for atom2 in atoms_H:
        distances = atoms.get_distances(atom2, atoms_H_bond)
        targets = np.delete(distances, np.where(distances >= HBond))
        targets.sort()
        pre_angle_group = []
        
        if len(targets) == 2:
            if targets[0] >= 1.2:
                continue
            for i, atom in enumerate(atoms_H_bond):
                distance = atoms.get_distance(atom, atom2)
                if distance in targets:
                    if atom not in angle_group:
                        pre_angle_group.append(atom)
                    pre_angle_group.append(atom2)
            if atoms.get_angle(pre_angle_group[0], pre_angle_group[1],\
                               pre_angle_group[2]) >= 150:
                angle_group.append(pre_angle_group[:3])
                
    return angle_group

file = 'Iimide2H_0.000'
angle_group = HBondNum(file, HBond=2.2)
print(angle_group)
print(len(angle_group))
```

```python
def near(atomnum, atoms, tolerance=2.000):
    atoms_not_H = []
    for i, atom in enumerate(atoms):
        if atom.symbol != 'H':
            atoms_not_H.append(i)
    mole = [atomnum]
    pre_distances = atoms.get_distances(atomnum, atoms_not_H)
    distances = np.delete(pre_distances, np.where(pre_distances >= tolerance))
    distances = np.delete(distances, np.where(distances == 0))
    distances.sort()
    length = len(distances)
    if length > 4:
        distances = distances[:4]
    for i, atom in enumerate(atoms_not_H):
        distance = atoms.get_distance(atom, atomnum)
        if distance in distances:
            mole.append(atom)
    return mole

def nearH(atomnum, atoms, tolerance=1.200):
    atoms_H = []
    for i, atom in enumerate(atoms):
        if atom.symbol == 'H':
            atoms_H.append(i)
    mole = [atomnum]
    pre_distances = atoms.get_distances(atomnum, atoms_H)
    distances = np.delete(pre_distances, np.where(pre_distances >= tolerance))
    distances = np.delete(distances, np.where(distances <= 0.5))
    distances.sort()
    length = len(distances)
    if length > 3:
        distances = distances[:4]
    for i, atom in enumerate(atoms_H):
        distance = atoms.get_distance(atom, atomnum)
        if distance in distances:
            mole.append(atom)
    return mole

def smileorder(file, loc='./', atomnum=0, tolerance=2.000, toleranceH=1.200,\
               form='XYZ', must_bonds={}):
    reverse = []
    if must_bonds != {}:
        for i,key in enumerate(must_bonds.keys()):
            for j,value in enumerate(must_bonds.values()):
                if i == j:
                    reverse.append([key, value])
    for couple in reverse:
        must_bonds[couple[1]] = couple[0]
    opath = os.getcwd()
    os.chdir(loc)
    if form == 'XYZ':
        atoms = read(file+'.xyz')
    elif form == 'LAMMPSDATA':
        DataToXyz(file, savename = 'soload')
        atoms = read('soload.xyz')
    atoms_not_H = []
    atoms_H = []
    for i, atom in enumerate(atoms):
        if atom.symbol != 'H':
            atoms_not_H.append(i)
        else:
            atoms_H.append(i)
    mole = near(atomnum, atoms, tolerance=tolerance)
    if atomnum in must_bonds.keys():
        mole.append(must_bonds[atomnum])
    i = 1
    while 1:
        molefrag = near(mole[i], atoms, tolerance=tolerance)
        if mole[i] in must_bonds.keys():
            molefrag.append(must_bonds[mole[i]])
        for num in molefrag:
            if num not in mole:
                mole.append(num)
        i = i + 1
        try:
            mole[i] == mole[-1]
        except:
            break
    fullmole = []
    for num in mole:
        fullmole.append(num)
        moleHfrag = nearH(num, atoms, tolerance=toleranceH)
        for num2 in moleHfrag:
            if num2 not in mole:
                fullmole.append(num2)
    if form == 'LAMMPSDATA':
        os.system('rm soload.xyz')
    os.chdir(opath)
    return fullmole
```

```python
import matplotlib.pyplot as plt
from ase.io import read
from ase import Atoms

def crossover_num(file, atomnum, screen='z', shadow_tolerance=1.25, height_tolerance=3.5):
    atoms = read(file+'.xyz')
    
    mole = molesplit(file, atomnum=atomnum[0])
    mole2 = molesplit(file, atomnum=atomnum[1])
    mole_not_H_x = []
    mole_not_H_y = []
    mole_not_H_z = []
    mole2_not_H_x = []
    mole2_not_H_y = []
    mole2_not_H_z = []
    linkers = []
    linker_pos = []
    crossover = 0
    
    for i in mole:
        mole_not_H_x.append(atoms[i].position[0])
        mole_not_H_y.append(atoms[i].position[1])
        mole_not_H_z.append(atoms[i].position[2])
    
    for i in mole2:
        mole2_not_H_x.append(atoms[i].position[0])
        mole2_not_H_y.append(atoms[i].position[1])
        mole2_not_H_z.append(atoms[i].position[2])
    
    if screen in ('z', 'Z'):
        linkers.append([0,0])
        if mole_not_H_z[0] - mole2_not_H_z[0] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
        
        for i1, x1 in enumerate(mole_not_H_x):
            for i2, x2 in enumerate(mole2_not_H_x):
                if ((mole_not_H_x[i1] - mole2_not_H_x[i2]) ** 2
                    + (mole_not_H_y[i1] - mole2_not_H_y[i2]) ** 2 <= shadow_tolerance ** 2):
                    if abs(mole_not_H_z[i1] - mole2_not_H_z[i2]) >= height_tolerance:
                        linkers.append([i1, i2])
                        if mole_not_H_z[i1] - mole2_not_H_z[i2] > 0:
                            linker_pos.append(1)
                        else:
                            linker_pos.append(-1)
                        break
    
        linkers.append([len(mole)-1, len(mole2)-1])
        if mole_not_H_z[-1] - mole2_not_H_z[-1] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
    elif screen in ('y', 'Y'):
        linkers.append([0,0])
        if mole_not_H_y[0] - mole2_not_H_y[0] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
        
        for i1, x1 in enumerate(mole_not_H_x):
            for i2, x2 in enumerate(mole2_not_H_x):
                if ((mole_not_H_x[i1] - mole2_not_H_x[i2]) ** 2
                    + (mole_not_H_z[i1] - mole2_not_H_z[i2]) ** 2 <= shadow_tolerance ** 2):
                    if abs(mole_not_H_y[i1] - mole2_not_H_y[i2]) >= height_tolerance:
                        linkers.append([i1, i2])
                        if mole_not_H_y[i1] - mole2_not_H_y[i2] > 0:
                            linker_pos.append(1)
                        else:
                            linker_pos.append(-1)
                        break
    
        linkers.append([len(mole)-1, len(mole2)-1])
        if mole_not_H_y[-1] - mole2_not_H_y[-1] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
    elif screen in ('x', 'X'):
        linkers.append([0,0])
        if mole_not_H_x[0] - mole2_not_H_x[0] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
        
        for i1, x1 in enumerate(mole_not_H_x):
            for i2, x2 in enumerate(mole2_not_H_x):
                if ((mole_not_H_y[i1] - mole2_not_H_y[i2]) ** 2
                    + (mole_not_H_z[i1] - mole2_not_H_z[i2]) ** 2 <= shadow_tolerance ** 2):
                    if abs(mole_not_H_x[i1] - mole2_not_H_x[i2]) >= height_tolerance:
                        linkers.append([i1, i2])
                        if mole_not_H_x[i1] - mole2_not_H_x[i2] > 0:
                            linker_pos.append(1)
                        else:
                            linker_pos.append(-1)
                        break
    
        linkers.append([len(mole)-1, len(mole2)-1])
        if mole_not_H_x[-1] - mole2_not_H_x[-1] > 0:
            linker_pos.append(1)
        else:
            linker_pos.append(-1)
        
    for i, stat in enumerate(linker_pos):
        try:
            if i > 0:
                if linker_pos[i] != linker_pos[i+1] and linker_pos[i] != linker_pos[i-1]:
                    linker_pos[i] = linker_pos[i-1]
                continue
        except:
            pass
    for i, stat in enumerate(linker_pos):
        try:
            if linker_pos[i] != linker_pos[i+1]:
                crossover = crossover + 1
        except:
            pass
    print(linkers, linker_pos)
    return crossover    
    
file = 'M3A3_0.140'
crossovers = crossover_num(file, atomnum=[3,16], screen='y', shadow_tolerance=1, height_tolerance=2)
print(crossovers)
```

```python
## from ase.io import read
from ase import Atoms
import numpy as np
import matplotlib.pyplot as plt

def neardraw(atomnum, atoms, tolerance=2.000, colour='b', plane='z'):
    atoms_not_H = []
    for i, atom in enumerate(atoms):
        if atom.symbol != 'H':
            atoms_not_H.append(i)
    mole = [atomnum]
    pre_distances = atoms.get_distances(atomnum, atoms_not_H)
    distances = np.delete(pre_distances, np.where(pre_distances >= tolerance))
    distances = np.delete(distances, np.where(distances == 0))
    distances.sort()
    length = len(distances)
    if length > 4:
        distances = distances[:4]
    for i, atom in enumerate(atoms_not_H):
        distance = atoms.get_distance(atom, atomnum)
        pos = atoms.get_positions()
        if distance in distances:
            if plane == 'z':
                plt.plot([pos[atom][0], pos[atomnum][0]],\
                         [pos[atom][1], pos[atomnum][1]], colour)
            elif plane == 'y':
                plt.plot([pos[atom][0], pos[atomnum][0]],\
                         [pos[atom][2], pos[atomnum][2]], colour)
            elif plane == 'x':
                plt.plot([pos[atom][1], pos[atomnum][1]],\
                         [pos[atom][2], pos[atomnum][2]], colour)
            mole.append(atom)
    return mole

def moleshadow(file, atomnum=0, tolerance=2.000, withH=False, colour='b', plane='z'):
    atoms = read(file+'.xyz')
    atoms_not_H = []
    atoms_H = []
    for i, atom in enumerate(atoms):
        if atom.symbol != 'H':
            atoms_not_H.append(i)
        else:
            atoms_H.append(i)
    mole = near(atomnum, atoms, tolerance=tolerance)
    i = 1
    while 1:
        molefrag = neardraw(mole[i], atoms, tolerance=tolerance, colour=colour, plane=plane)
        for num in molefrag:
            if num not in mole:
                mole.append(num)
        i = i + 1
        try:
            mole[i] == mole[-1]
        except:
            break
    if withH == True:
        mole_H = []
        for atom in mole:
            pre_distances = atoms.get_distances(atom, atoms_H)
            distances = np.delete(pre_distances, np.where(pre_distances >= 1.2))
            for H in atoms_H:
                distance = atoms.get_distance(H, atom)
                if distance in distances:
                    mole_H.append(H)
        mole = mole + mole_H
    return mole
```

```python
import re
import warnings

def XyzToPdb_Multi(file, size, steps=[], pattern=[90, 90, 90], loc='./', saveloc='./', savename=''):
    step = 1
    f = open(loc+file+'.xyz', 'r')
    if savename == '':
        savename = file
    try:
        w = open(saveloc+savename+'.pdb', 'x')
    except:
        w = open(saveloc+savename+'.pdb', 'w')
    stepnum = 1
    i = 0
    xmin = 99999
    xmax = -99999
    ymin = 99999
    ymax = -99999
    zmin = 99999
    zmax = -99999
    w.write('AUTHOR    MCPoly\n')
    j = 0
    for line in f:
        if j == 0:
            num = eval(line[:-1])
        if j in [0, 1]:
            j = j + 1
            continue
        b = re.findall(r'\-?[0-9]+\.[0-9]+', line)
        if len(b) == 3:
            if eval(b[0]) < xmin:
                xmin = eval(b[0])
            if eval(b[0]) > xmax:
                xmax = eval(b[0])
            if eval(b[1]) < ymin:
                ymin = eval(b[1])
            if eval(b[1]) > ymax:
                ymax = eval(b[1])
            if eval(b[2]) < zmin:
                zmin = eval(b[2])
            if eval(b[2]) > zmax:
                zmax = eval(b[2])
        elif len(b) == 0:
            stepnum = stepnum + 1
            j = 0
        j = j + 1
    j = 0
    f.close()
    f = open(loc+file+'.xyz', 'r')
    for line in f:
        if j == 1:
            if steps == [] or (step in steps) or (step-stepnum-1 in steps):
                w.write('REMARK    Step {0}\n'.format(step))
                w.write('CRYST1  {0:>7.3f}  {1:>7.3f}  {2:>7.3f}  {3:.2f}  {4:.2f}  {5:.2f}\n'.format(*size, *pattern))
        if j in [0, 1]:
            j = j + 1
            continue
        ele = re.search(r'[A-Z][a-z]*', line)
        b = re.findall(r'\-?[0-9]+\.[0-9]+', line)
        #print(ele,b)
        if ele and len(b) == 3:
            if eval(b[0])+size[0]/2-(xmax+xmin)/2 < 0.0:
                warnings.warn("Some atoms are out of the cell box.")
            elif eval(b[1])+size[1]/2-(ymax+ymin)/2 < 0.0:
                warnings.warn("Some atoms are out of the cell box.")
            elif eval(b[2])+size[2]/2-(zmax+zmin)/2 < 0.0:
                warnings.warn("Some atoms are out of the cell box.")
            elif eval(b[0])+size[0]/2-(xmax+xmin)/2 > size[0]:
                warnings.warn("Some atoms are out of the cell box.")
            elif eval(b[1])+size[1]/2-(ymax+ymin)/2 > size[1]:
                warnings.warn("Some atoms are out of the cell box.")
            elif eval(b[2])+size[2]/2-(zmax+zmin)/2 > size[2]:
                warnings.warn("Some atoms are out of the cell box.")
            if steps == [] or (step in steps) or (step-stepnum-1 in steps):
                w.write('ATOM  {0:>5} {1:<2}                 '.format(j-1, ele.group(0)))
                w.write('{0:>7.3f} {1:>7.3f} {2:>7.3f} '.format(eval(b[0])+size[0]/2-(xmax+xmin)/2, eval(b[1])+size[1]/2-(ymax+ymin)/2, eval(b[2])+size[2]/2-(zmax+zmin)/2))
                w.write('{0:>5.2f} {1:>5.2f}           {2}\n'.format(0.00, 0.00, ele.group(0)))
                if j == num + 1:
                    w.write('END\n')
        elif len(b) == 0:
            step = step + 1
            j = 0
        j = j + 1
    f.close()
    i = 0
    w.close()
```
